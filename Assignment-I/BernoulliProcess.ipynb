# HIV Virus Transmission
<br/>

$\text{The process is both discrete state and discrete time and can be modelled as a Bernoulli}$ 
$\text{Random Process with the following characteristics:}$ 
+ $\text{Every transfusion represents a unit time interval.}$
+ $\text{Each instance of transfusion represents a Bernoulli Trial in which the}$ 
$\text{probability of the virus getting transferred successfully is p.}$

$\text{Formally, if $i$ represents $i^{th}$ interval in the process}$

 $T_{i} = \left\{
    \begin{array}{ll}
          0 & \text{failure of virus transmission} \\
          1 & \text{success of virus transmission} \\
    \end{array} 
    \right. \forall i \geqslant 1$ 
        
$\text{P($T_{0}$=1) = 0: Arrivals start after t = 0.}$ 

$\text{P($T_{i}$>1) = 0: There can't be multiple arrivals in a unit interval.}$ 

$\text{P($T_{i}$=1) = p}$ 

$\text{Each } T_{i} \text{ represents an independent Bernoulli Trial.}$


$\text{(a) For a patient with p = 0.8, we generate the process for a finite time t = 500 units.}$
$\text{The x-axis represents the time where unit time = time elapsed during one transfusion.}$
$\text{The y-axis represents the process where 1 represents success of a trial whereas 0}$ 
$\text{represents a failure.}$




```r
p <- 0.8
process <- rbinom(500,1,p)
t <- 1:500
plot(t, process, main=paste("Number of Successful Trials =",sum(process)))
```

$\text{(b) First, we derive the distribution for the interarrival times.}$
$\text{Since each trial is independent of the other, the probability of success for every trial}$
$\text{remains p. Let's say we encounter an arrival or a success in the process and prior to}$ 
$\text{the next arrival, there are k-1 failures.}$
$\text{Therefore, } P(X_{i}=k) =
P(T_{1}=0 \land T_{2}=0 .... T_{k-1}=0 \land  T_{k}=1)$
$\text{= } P(T_{1}=0)P(T_{2}=0)....P(T_{k-1}=0)P(T_{k}=1) = (1-p)^{k-1}p$
$\text{Hence, } $$X_{i}$$ \text{ ~ geometric(p)}$
<br/>
$\text{For this question, we plot the distribution of }$ $X_{1}$ $\text{till t = 30 with p = 0.3}$

```r
F = pgeom(1:30, 0.3)
X = 1:30
plot(X, F, xlab='X', ylab='F(X)')
```

$\text{(c) First, we plot the distribution for } X_{1} = 1000 \text{ till t = 1000 with p = 0.01}$ 

```r
F = pgeom(1:1000, 0.01)
X = 1:1000
plot(X, F, xlab='X', ylab='F(X)')
```

$\text{We observe that, with the increase in time slots, the inter-arrival times depict an exponential}$
$\text{behavior. This is because they are geometric in nature and geometric random variables}$
$\text{exhibit an exponential behavior as the number of instances grow infinitely large or}$
$\text{the lengths of the intervals become infinitesimally small.}$

$\text{(d) This is a case of merging Bernoulli Processes where the processes of 3 independent}$
$\text{patients are getting merged. So we model it in a way such that at each Bernoulli trial}$
$\text{involves 3 transfusions and there are 4 possibilities in terms of arrivals.}$

$\text{Formally, if $i$ represents $i^{th}$ interval in the process}$

 $T_{i} = \left\{
    \begin{array}{ll}
          0 & \text{no virus transmission takes place} \\
          1 & \text{virus transmission takes place for 1 patient} \\
          2 & \text{virus transmission takes place for 2 patients} \\
          3 & \text{virus transmission takes place in all (3) patients} \\
    \end{array} 
    \right. \forall i \geqslant 1$ 
    
$\text{Since the patients are independent and the probability of success is same for all of them,}$
$\text{we can treat each of the}$ $T_{i}'s$ $\text{in the three processes as iid's.}$ 
$\text{The probabilities for each of the above cases can be written as:}$
<br/>
$P(T_{i}=0) = {3\choose 0}(1-p)^{3}$
<br/>
$P(T_{i}=1) = {3\choose 1}(1-p)^{2}p$
<br/>
$P(T_{i}=2) = {3\choose 2}(1-p)p^{2}$
<br/>
$P(T_{i}=3) = {3\choose 3}p^{3}$

$\text{Now, we go about plotting the number of transfusions in 50 Bernoulli Trials with}$
$\text{varying probabilities for p = 0.4, 0.5, 0.6, 0.7, 0.8}$


```r
arrivals_4 <- rbinom(50,3,0.4)
arrivals_5 <- rbinom(50,3,0.5)
arrivals_6 <- rbinom(50,3,0.6)
arrivals_7 <- rbinom(50,3,0.7)
arrivals_8 <- rbinom(50,3,0.8)

sum_4 = sum(arrivals_4)
sum_5 = sum(arrivals_5)
sum_6 = sum(arrivals_6)
sum_7 = sum(arrivals_7)
sum_8 = sum(arrivals_8)

print(paste("# of HIV transmissions for p=0.4:", sum_4))
print(paste("# of HIV transmissions for p=0.5:", sum_4))
print(paste("# of HIV transmissions for p=0.6:", sum_6))
print(paste("# of HIV transmissions for p=0.7:", sum_7))
print(paste("# of HIV transmissions for p=0.8:", sum_8))

for(i in 2:50){
    arrivals_4[i] <- arrivals_4[i-1] + arrivals_4[i]
    arrivals_5[i] <- arrivals_5[i-1] + arrivals_5[i]
    arrivals_6[i] <- arrivals_6[i-1] + arrivals_6[i]
    arrivals_7[i] <- arrivals_7[i-1] + arrivals_7[i]
    arrivals_8[i] <- arrivals_8[i-1] + arrivals_8[i]
}

x <- 1:50
plot(x,arrivals_4, col='red', type='o', xlab = 'occurences', ylab = 'no. of HIV transmissions')
lines(x, arrivals_5, col='blue', type='o')
lines(x, arrivals_6, col='green', type='o')
lines(x, arrivals_7, col='black', type='o')
lines(x, arrivals_8, col='violet', type='o')
```

$\text{We observe that the number of arrivals or HIV transmissions are greater for a higher}$
$\text{value of p. The CDF of the process with a higher p grows faster than others with respect}$
$\text{to the number of occurences or transfusions.}$
